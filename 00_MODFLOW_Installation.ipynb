## Install mfpymake: building MODFLOW-based programs from source files.

We use mfpymake (https://github.com/modflowpy/pymake) USGS' Python package for compiling MODFLOW-based programs.

```python
# Install python package/module: mfpymake

!pip install mfpymake==1.2.10
```

```python
import pymake
```

Unfortunately, there is an installation issue in Colab with sys.argv (see below in the `previous implementation` section) since pymake version 1.2.0. We need to change sys.argv temporarily (or see below without sys.argv modification)

```python
import sys
```

```python
print(sys.argv)
myargv = sys.argv
```

Current pymake version checks the arguments when sys.argv[0] contains '.py' to see whether pymake runs in command line. This would cause error when you run this script in Colab notebook.

We bypass this issue by assigning sys.argv to a dummy string (do not use ".py" or it will produce error) and restore the values later

```python
sys.argv = ['ipython']
```

## Install MODFLOW 6: the newest version of MODFLOW

```python
! gfortran --version
```

```python
pymake.build_apps('mf6')
```

reset `sys.argv`

```python
sys.argv = myargv
print(sys.argv)
```

check if we have successfully installed MODFLOW 6

```python
!ls -al
```

Now we run mf6 - we just run mf6 without inputs and you need to stop instruction

```python
# run modflow 6

!./mf6 # of course mf6 waits for your inputs since we didn't provide NAM file. Ctrl + M I to stop/interrupt instruction
```

if `!./mf6` was not working, it is probably due to permission problem. Change the permission as below:

```python
!chmod 755 mf6
```

Copy your executable to relevant locations. You can copy it to /usr/local/bin/ use it without specifying its full location (/usr/local/bin/mf6) because /usr/local/bin/ is in the PATH

```python
!cp ./mf6 /usr/local/bin
```

```python
# run modflow 6
!mf6 # stop this again
```

The name of resulting executable is "mf6" so please use this executable for MODFLOW 6 Flopy applications.
You may want to copy "mf6" to your drive otherwise you need to compile mf6 everytime you restart the runtime.
